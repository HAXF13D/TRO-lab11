### Медианный фильтр

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Функция внесения шума соли и перца на изображение

```python
def add_salt_noise(img, snr):
    size = img.size
    noiseSize = int(size * (1 - snr))
    for k in range(0, noiseSize):
        xi = int(np.random.uniform(0, img.shape[1]))
        xj = int(np.random.uniform(0, img.shape[0]))
        if img.ndim == 2:
            img[xj, xi] = 255
        elif img.ndim == 3:
            img[xj, xi] = 0
    return img
```

### Вносим шум на изображение

```python
filename = "1.jpg"
img = add_salt_noise(cv2.imread(filename), 0.95)

rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(rgb_img) 
```

#### Медианный фильтр с размером ядра = 3

```python
lbimg = cv2.medianBlur(img, 3)

rgb_img = cv2.cvtColor(lbimg,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(rgb_img) 
```
